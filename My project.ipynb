```python
import pandas as pd
import altair as alt
df= pd.read_csv("Coffee Sales index_New.csv")
df.head()
```

```python
#Distribution of coffees by Coffee type -- BAR  ***************
df['date'] = pd.to_datetime(df['date1'], format='%B')

# Calculate distribution of coffee by cups
coffee_distribution = df.groupby('coffee_name').size().reset_index(name='cups_sold')

# Create Altair bar chart
alt.Chart(coffee_distribution).mark_bar().encode(
    x='coffee_name',
    y='cups_sold',
    color = 'coffee_name'
).properties(
    title='Distribution of coffees by Coffee type',
    width=700,
    height=300)
```

```python
#Revenue generated each month - Line

alt.Chart(df).mark_line().encode(
    x=alt.X('date1', title='Month'),
    y=alt.Y('sum(money)', title='Revenue'),
    color = 'coffee_name'
).properties(
    title='Revenue for each month - Line',
    width=700,
    height=300)
```

```python
#Revenue generated each month - bar

alt.Chart(df).mark_bar().encode(
    x=alt.X('date1', title='Month'),
    y=alt.Y('sum(money)', title='Revenue'),
    #color = 'coffee_name'
    color=alt.Color('coffee_name', scale=alt.Scale(scheme='spectral')),
).properties(
    title='Revenue for each month - Bar',
    width=700,
    height=300)
```

```python
#Distribution of Coffees by datetime

alt.Chart(df).mark_circle().encode(
    x=alt.X('datetime', title='Month'),
    y=alt.Y('coffee_name', title='coffee_name'),
    color=alt.Color('date1', scale=alt.Scale(scheme='spectral')),
    tooltip=['date1', 'datetime']
    #color = 'date'
).properties(
    title='Coffee types sold vs datetime',
    width=700,
    height=300)
```

```python
#Revenue generated by datetime
alt.Chart(df).mark_circle().encode(
    x=alt.X('datetime', title='datetime'),
    y=alt.Y('sum(money)', title='Revenue'),
    color = 'coffee_name'
).properties(
    title='Revenue by datetime',
    width=700,
    height=300)
```

```python
#Revenue by coffee type -- bar

a = alt.Chart(df).mark_bar().encode(
    x=alt.X('coffee_name', title='coffee_name'),
    y=alt.Y('sum(money)', title='Revenue'),
    color = 'coffee_name'
).properties(
    title='Revenue by coffee type',
    width=400,
    height=300)

#Revenue at different times

b = alt.Chart(df).mark_bar().encode(
    x=alt.X('date1', title='Month'),
    y=alt.Y('sum(money)', title='Revenue'),
    color = 'coffee_name'
).properties(
    title='Revenue at different times',
    width=300,
    height=300)

a | b
```

```python
#Distribution of Coffee Payment Types  **********************

# Create a bar chart showing the total money spent on each coffee type
alt.Chart(df).mark_bar().encode(
    x=alt.X('cash_type', title='Coffee Type'),
    y=alt.Y('count(cash_type)', title='Count of payment_type'),
    color = 'coffee_name'
).properties(
    title='Distribution of Coffee Payment Types',
    width=700,
    height=300)
```
